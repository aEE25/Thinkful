### Determine the five most common journals and the total articles for each. 

```python
#make sure all Journal titles are entered as strings
wellcome['Journal title'] = wellcome['Journal title'].astype(str)
```

### Next, calculate the mean, median, and standard deviation of the open-access cost per article for each journal . 

```python
#remove pount sign from cost
def RemovePound(val):
    if val[0] == '£':
        number = val.split('£')[1]
        return number
    else:
        return val
    
wellcome['cost'] = wellcome['cost'].apply(RemovePound)
```

```python
#change cost from string to number
wellcome['cost'] = pd.to_numeric(wellcome['cost'], errors='coerce')
```
