### Prepare data for modeling

```python
countries.isnull().sum(axis=0)
```

```python
countries_dropped = countries.dropna()
```

```python
Y = countries_dropped['Net migration']
X = countries_dropped.drop(['Net migration', 'Country', 'Region'], 1)
```

```python
def model_results(model):
    model.fit(X, Y)
    score = cross_val_score(model,X,Y, cv=5)
    y_pred = cross_val_predict(model,X,Y,cv=5)
    print('Cross Validation Scores:\n', score)
    print('\nR-squared:')
    print(model.score(X, Y))
```

```python
def model_residuals(model):
    predicted = model.predict(X).ravel()
    residual = Y - predicted
    
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.hist(residual)
    plt.title('Residual counts')
    plt.xlabel('Residual')
    plt.ylabel('Count')
    
    plt.subplot(1, 3, 2)
    plt.scatter(predicted, residual)
    plt.xlabel('Predicted')
    plt.ylabel('Residual')
    plt.axhline(y=0)
    plt.title('Residual vs. Predicted')
    
    plt.subplot(1, 3, 3)
    plt.scatter(predicted, Y)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.axhline(y=0)
    plt.title('Actual vs. Predicted')
    plt.show()
    
    plt.show()
```

## Ridge Regression

```python
# tune ridge

r_results = pd.DataFrame()

def lambda_test(value):
    #Test lambda value in ridge model
    ridge = linear_model.Ridge(alpha=value, fit_intercept=False)  
    ridge_score = cross_val_score(ridge, X, Y, cv=5)
    #return dataframe of r-squared for each lambda value
    data = pd.DataFrame({"lambda_value": [value], "ridge_score": [ridge_score.mean()], "std": [ridge_score.std()]})
    return data

#Test large values for lamdba
appended_data = []
lambda_range = np.arange(5, 20, 0.5)
for i in lambda_range:
    result = lambda_test(i)
    appended_data.append(result)
ridge_results = pd.concat(appended_data, ignore_index=True)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.plot(ridge_results['lambda_value'], ridge_results['ridge_score'], label="R-Squared")
plt.xlabel('Lambda Value')
plt.ylabel('R-squared Value')
plt.legend(loc='best') 
plt.title('Ridge Lambda Test')

plt.subplot(1, 2, 2)
plt.plot(ridge_results['lambda_value'], ridge_results['std'], label="STD")
plt.xlabel('Lambda Value')
plt.ylabel('STD Value')
plt.legend(loc='best') 
plt.title('Ridge Lambda Test')

plt.show()

print('Ridge Lambda Value(s) with lowest STD:')
ridge_results.loc[ridge_results['std'] == ridge_results['std'].min()]
```

```python
ridge = linear_model.Ridge(alpha=19.5, fit_intercept=False) 
```

```python
model_results(ridge)
```

```python
model_residuals(ridge)
```

# Can we predict whether a country has net migration gain?

Establish of countries with net migration gain by creating a variable class.

```python
countries_dropped['migration_gain'] = 0
countries_dropped.loc[countries_dropped['Net migration'] > 0,'migration_gain'] = 1
```

```python
plt.figure(figsize=(5, 5))
plt.title('Migration Class Counts')
pd.value_counts(countries_dropped['migration_gain']).plot.bar()
plt.xticks(rotation = 0)
plt.show()

print(countries_dropped['migration_gain'].value_counts())

print("\nPortion of countries with Net Migration Gain: %0.3f" % (countries_dropped['migration_gain'].value_counts()[1]/countries_dropped.shape[0]))
```

Almost 27% of countries show net migration gain.  Let's try some models to improve on that baseline.

```python
Y_class = countries_dropped['migration_gain']
X_class = countries_dropped.drop(['Net migration', 'Country', 'Region', 'migration_gain'], 1)
```

```python
def class_model_results(model):
    score = cross_val_score(model,X_class,Y_class, cv=5)
    y_pred = cross_val_predict(model,X_class,Y_class,cv=5)
    print('Cross Validation Scores:\n', score)
    print("\nCross Validation Range: %0.3f to %0.3f" % (score.min(), score.max()))
    print("\nCross Validation Average: %0.3f" % (score.mean()))
```

```python
#Test n_estimators

forest_results = pd.DataFrame()

def forest_tree_fit(tree):
    rfc = ensemble.RandomForestClassifier(
        n_estimators=tree
    )
    forest_score = cross_val_score(rfc, X_class, Y_class, cv=5)
    
    data = pd.DataFrame({"tree_number": [tree], "score": [forest_score.mean()], "std": [forest_score.std()]})
    return data
                                       
appended_data = []
test_range = np.arange(150, 300, 10)
for i in test_range:
    result = forest_tree_fit(i)
    appended_data.append(result)
forest_results = pd.concat(appended_data, ignore_index=True)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.plot(forest_results['tree_number'], forest_results['score'], label="R-Squared")
plt.xlabel('Estimator Number')
plt.ylabel('R-squared Value')
plt.legend(loc='best') 
plt.title('Forest Estimator Test')

plt.subplot(1, 2, 2)
plt.plot(forest_results['tree_number'], forest_results['std'], label="STD")
plt.xlabel('Estimator Number')
plt.ylabel('STD Value')
plt.legend(loc='best') 
plt.title('Forest Estimator Test')

plt.show()

print('Forest Estimator Number with lowest STD:')
forest_results.loc[forest_results['std'] == forest_results['std'].min()]
```

```python
#Test max_features

forest_results = pd.DataFrame()

def forest_feature_fit(feature):
    rfc = ensemble.RandomForestClassifier(
        n_estimators=270, max_features=feature
    )
    forest_score = cross_val_score(rfc, X_class, Y_class, cv=5)
    
    data = pd.DataFrame({"feature_number": [feature], "score": [forest_score.mean()], "std": [forest_score.std()]})
    return data
                                       
appended_data = []
test_range = np.arange(1, 17, 1)
for i in test_range:
    result = forest_feature_fit(i)
    appended_data.append(result)
forest_results = pd.concat(appended_data, ignore_index=True)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.plot(forest_results['feature_number'], forest_results['score'], label="R-Squared")
plt.xlabel('Feature Number')
plt.ylabel('R-squared Value')
plt.legend(loc='best') 
plt.title('Forest Feature Test')

plt.subplot(1, 2, 2)
plt.plot(forest_results['feature_number'], forest_results['std'], label="STD")
plt.xlabel('Feature Number')
plt.ylabel('STD Value')
plt.legend(loc='best') 
plt.title('Forest Feature Test')

plt.show()

print('Forest Feature Number with lowest STD:')
forest_results.loc[forest_results['std'] == forest_results['std'].min()]
```

```python
#Test max_depth

forest_results = pd.DataFrame()

def forest_depth_fit(depth):
    rfc = ensemble.RandomForestClassifier(
        n_estimators=270, max_features=4, max_depth=depth
    )
    forest_score = cross_val_score(rfc, X_class, Y_class, cv=5)
    
    data = pd.DataFrame({"depth_number": [depth], "score": [forest_score.mean()], "std": [forest_score.std()]})
    return data
                                       
appended_data = []
test_range = np.arange(1, 50, 2)
for i in test_range:
    result = forest_depth_fit(i)
    appended_data.append(result)
forest_results = pd.concat(appended_data, ignore_index=True)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.plot(forest_results['depth_number'], forest_results['score'], label="R-Squared")
plt.xlabel('Depth Number')
plt.ylabel('R-squared Value')
plt.legend(loc='best') 
plt.title('Forest Depth Test')

plt.subplot(1, 2, 2)
plt.plot(forest_results['depth_number'], forest_results['std'], label="STD")
plt.xlabel('Depth Number')
plt.ylabel('STD Value')
plt.legend(loc='best') 
plt.title('Forest Depth Test')

plt.show()

print('Forest Feature Number with lowest STD:')
forest_results.loc[forest_results['std'] == forest_results['std'].min()]
```
