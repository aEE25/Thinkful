### Lasso Feature Selection

```python
# tune lasso

r_results = pd.DataFrame()

def lambda_test(value):
    #Test lambda value in lasso model
    lass = linear_model.Lasso(alpha=value) 
    lass_score = cross_val_score(lass, X, Y, cv=10)
    #return dataframe of r-squared for each lambda value
    data = pd.DataFrame({"lambda_value": [value], "lass_score": [lass_score.mean()], "std": [lass_score.std()]})
    return data

#Test large values for lamdba
appended_data = []
lambda_range = np.arange(1e6, 1e8, 1e6)
for i in lambda_range:
    result = lambda_test(i)
    appended_data.append(result)
lasso_results = pd.concat(appended_data, ignore_index=True)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.plot(lasso_results['lambda_value'], lasso_results['lass_score'], label="R-Squared")
plt.xlabel('Lambda Value')
plt.ylabel('R-squared Value')
plt.legend(loc='best') 
plt.title('Lasso Lambda Test')

plt.subplot(1, 2, 2)
plt.plot(lasso_results['lambda_value'], lasso_results['std'], label="STD")
plt.xlabel('Lambda Value')
plt.ylabel('STD Value')
plt.legend(loc='best') 
plt.title('Lasso Lambda Test')

plt.show()

print('Lasso Lambda Value(s) with lowest STD:')
lasso_results.loc[lasso_results['std'] == lasso_results['std'].min()]
```

### Result Metrics

```python
def pca_model_results(model):
    start_time = time.time()
    score = cross_val_score(model,X_pca,Y, cv=5)
    print("Cross Validation Scores", score)  
    model.fit(X_pca, Y)
    print('\nR-squared: ', model.score(X_pca, Y))
    print("\n--- %s seconds ---" % (time.time() - start_time))
```

```python
def pca_model_residuals(model):
    predicted = model.predict(X_pca).ravel()
    residual = Y - predicted
    
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.hist(residual)
    plt.title('Residual counts')
    plt.xlabel('Residual')
    plt.ylabel('Count')
    
    plt.subplot(1, 3, 2)
    plt.scatter(predicted, residual)
    plt.xlabel('Predicted')
    plt.ylabel('Residual')
    plt.axhline(y=0)
    plt.title('Residual vs. Predicted')
    
    plt.subplot(1, 3, 3)
    plt.scatter(predicted, Y)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.axhline(y=0)
    plt.title('Actual vs. Predicted')
    plt.show()
    
    plt.show()
```

```python
def model_results(model):
    start_time = time.time()
    score = cross_val_score(model,X,Y, cv=5)
    print("Cross Validation Scores", score)    
    model.fit(X, Y)
    print('\nR-squared: ', model.score(X, Y))
    print("\n--- %s seconds ---" % (time.time() - start_time))
```

```python
def model_residuals(model):
    predicted = model.predict(X).ravel()
    residual = Y - predicted
    
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.hist(residual)
    plt.title('Residual counts')
    plt.xlabel('Residual')
    plt.ylabel('Count')
    
    plt.subplot(1, 3, 2)
    plt.scatter(predicted, residual)
    plt.xlabel('Predicted')
    plt.ylabel('Residual')
    plt.axhline(y=0)
    plt.title('Residual vs. Predicted')
    
    plt.subplot(1, 3, 3)
    plt.scatter(predicted, Y)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.axhline(y=0)
    plt.title('Actual vs. Predicted')
    plt.show()
    
    plt.show()
```

### Tuning Regression with Lasso

```python
#Test alpha values

tune_results = pd.DataFrame()

def lasso_fit(value):
    lasso = linear_model.Lasso(alpha=value)
    lasso_score = cross_val_score(lasso, X, Y, cv=5)
    data = pd.DataFrame({"alpha_number": [value], "cross_val": [lasso_score.mean()], "std": [lasso_score.std()]})
    return data
                                       
appended_data = []
test_range = np.arange(50, 400, 10)
for i in test_range:
    result = lasso_fit(i)
    appended_data.append(result)
tune_results = pd.concat(appended_data, ignore_index=True)

print(tune_results.loc[tune_results['std'] == tune_results['std'].min()])
print(tune_results.loc[tune_results['cross_val'] == tune_results['cross_val'].max()])

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(tune_results['alpha_number'], tune_results['std'])
plt.xlabel('Alpha Value')
plt.ylabel('STD')
plt.legend(loc='best') 
plt.title('Lasso Alpha Test, STD')

plt.subplot(1, 2, 2)
plt.plot(tune_results['alpha_number'], tune_results['cross_val'])
plt.xlabel('Alpha Value')
plt.ylabel('CV Mean')
plt.legend(loc='best') 
plt.title('Lasso Alpha Test, CV')

plt.show()
```

```python
#Test alpha value with fit intercept

tune_results = pd.DataFrame()

def lasso_fit(value):
    lasso = linear_model.Lasso(alpha=value, fit_intercept=True)
    lasso_score = cross_val_score(lasso, X, Y, cv=5)
    lasso.fit(X, Y)
    data = pd.DataFrame({"alpha_number": [value], "score": [lasso.score(X, Y)], "cross_val": [lasso_score.mean()], "std": [lasso_score.std()]})
    return data
                                       
appended_data = []
test_range = np.arange(500, 10000, 500)
for i in test_range:
    result = lasso_fit(i)
    appended_data.append(result)
tune_results = pd.concat(appended_data, ignore_index=True)

print(tune_results.loc[tune_results['std'] == tune_results['std'].min()])
print(tune_results.loc[tune_results['cross_val'] == tune_results['cross_val'].max()])
print(tune_results.loc[tune_results['score'] == tune_results['score'].max()])

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(tune_results['alpha_number'], tune_results['std'])
plt.xlabel('Alpha Value')
plt.ylabel('STD')
plt.legend(loc='best') 
plt.title('Lasso Alpha Test, STD')

plt.subplot(1, 3, 2)
plt.plot(tune_results['alpha_number'], tune_results['cross_val'])
plt.xlabel('Alpha Value')
plt.ylabel('CV Mean')
plt.legend(loc='best') 
plt.title('Lasso Alpha Test, CV')

plt.subplot(1, 3, 3)
plt.plot(tune_results['alpha_number'], tune_results['score'])
plt.xlabel('Alpha Value')
plt.ylabel('R-Squared')
plt.legend(loc='best') 
plt.title('Lasso Alpha Test, R-squared')

plt.show()
```
